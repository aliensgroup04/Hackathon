```python
#Merging data frames of cleaned dataset and coordinates
merged_df = pd.merge(df, df_c, on='listed_incity', how='left')
merged_df
```

```python
#importing libraries
from folium.plugins import MarkerCluster
import folium
from IPython.display import IFrame
```

```python
#Filtiring italian cuisines
italian_df = merged_df[merged_df['cuisines'].str.contains('Italian', case=False,na=False)]
italian_df
```

```python
#html code for task 2
map4 = folium.Map(location=[12.9716, 77.5946], zoom_start=12)

for i, row in italian_df.iterrows():
    lat = row['Latitude']
    lon = row['Longitude']
    
    # Only add marker if lat and lon are valid (not NaN)
    if pd.notna(lat) and pd.notna(lon):
        folium.Marker(
            location=[lat, lon],
            popup=f"{row['listed_incity']}<br>Cuisine: {row['cuisines']}",
            icon=folium.Icon(color='purple')
        ).add_to(map4)

map4.save('italian_spots_map.html')
```
