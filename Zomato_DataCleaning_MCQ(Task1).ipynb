```python
#13th question
# 1. Group by 'listed_incity' (area) and calculate mean cost and total votes
area_revenue_df = df.groupby('listed_incity').agg({
    'approx_costfor_two_people': 'mean',
    'votes': 'sum'
}).reset_index()

# 2. Rename columns
area_revenue_df.columns = ['area', 'avg_cost_for_two', 'total_votes']

# 3. Calculate potential revenue
area_revenue_df['potential_revenue'] = area_revenue_df['avg_cost_for_two'] * area_revenue_df['total_votes']

# 4. Sort by potential revenue
area_revenue_df = area_revenue_df.sort_values('potential_revenue', ascending=False).reset_index(drop=True)

print(area_revenue_df.head())
```

```python
#14th question
# Group by 'rest_type' and calculate the mean rating for each type
rating_by_type = df.groupby('rest_type').agg({'rate': 'mean'}).reset_index()

# Sort to see which restaurant type has the lowest average rating (potentially more complaints)
rating_by_type = rating_by_type.sort_values('rate', ascending=True)

rating_by_type.head(50)
```

```python
#15th question
# 1. Filter based on the given conditions
filtered_df = df[(df['rate'] > 4.2) & (df['votes'] > 500) & (df['online_order'] == 1)]

# 2. Group by locality and count the number of restaurants
area_counts = filtered_df.groupby('listed_incity').size().reset_index(name='restaurant_count')

# 3. Sort by the number of qualifying restaurants
area_counts = area_counts.sort_values('restaurant_count', ascending=False)

# 4. View the top areas
print(area_counts)
```
